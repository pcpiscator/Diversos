```python
from pyspark.sql import SparkSession

# Crie uma sessão Spark
spark = SparkSession.builder.appName('MySparkApp').getOrCreate()

print('Sessão Spark iniciada com sucesso!')
print(f'Versão do Spark: {spark.version}')
```

```python
# Leia o arquivo Excel usando pandas
pdf = pd.read_excel("/content/Voleibol_Meses.xlsx")

# Converta o DataFrame do pandas para um DataFrame do Spark
df = spark.createDataFrame(pdf)

# Selecione um subconjunto de características e filtre por 'Salario' > 0
filtered_df = df.select("Posicao", "Equipe", "Salario").filter("Salario > 0")

# Gere estatísticas de resumo
filtered_df.summary().show()
```

```python
from pyspark.sql.functions import col, sum

# Verifica a contagem de valores nulos por coluna
df.select([sum(col(c).isNull().cast("integer")).alias(c) for c in df.columns]).show()
```
