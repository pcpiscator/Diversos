```python
# Extração de sites com Python

import requests
from bs4 import BeautifulSoup
import os

os.system('cls')

url = "https://eventos.sp.senac.br"

# Faz a requisição ao site
requisicao = requests.get(url)

# Converte o HTML
site = BeautifulSoup(requisicao.text, "html.parser")

# Busca todas as tags <time>
dataT = site.find_all("time")

# Exibe os textos encontrados
for data in dataT:
    print(data.text.strip())
```
