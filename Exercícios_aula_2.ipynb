```python
import os
os.system('cls')

num01 = float(input ("digite o primeiro valor"))
num02 = float(input ("digite o segundo valor"))

soma = (num01 + num02)
print(f"A soma é = {soma:.2f}")
sub = (num01 - num02)
print(f"A diferença é = {sub:.2f}")
prod = (num01 * num02)
print(f"O produto é = {prod:.2f}")
div = (num01 / num02)
print(f"A razão é = {div:.2f}")
```
