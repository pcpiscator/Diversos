```python
# Lista para armazenar os alunos

alunos = []

# Função para cadastrar alunos
def cadastrar_alunos(nome, nota1, nota2, nota3, nota4):
    aluno = (nome, nota1, nota2, nota3, nota4)
    alunos.append(aluno)

# Cadastro de exemplo
cadastrar_alunos("Maria", 8.0, 7.5, 9.0, 6.5)
cadastrar_alunos("João", 5.0, 6.0, 4.5, 5.5)

# Exibir resultados
for aluno in alunos:
    nome = aluno[0]
    media = (aluno[1] + aluno[2] + aluno[3] + aluno[4]) / 4

    if media >= 7:
        situacao = "Aprovado"
    else:
        situacao = "Reprovado"

    print(f"{nome} - Média: {media:.2f} - {situacao}")
```
